```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('iris')
```

```python
df.shape
```

```python
df.info().pd.set_option('display.max_rows',None)
```

```python
plt.title("Distribution of Sepal Length")
sns.histplot(df['sepal_length'])
```

```python
print("Distribution of Sepal Width")
sns.histplot(df['sepal_width'],kde=True)
```

```python
print("Distribution of Petal Length")
sns.histplot(df['petal_length'],bins=6)
```

```python
print("Distribution of Petal Width")
sns.histplot(df['petal_width'],bins=6)
```

```python
print("Distribution of Sepal Length")
sns.boxplot( x="sepal_length",data=df,)
```

```python
print("Distribution of Sepal Width")
sns.boxplot( y="sepal_width",data=df,)
```

```python
print("Distribution of Petal Length")
sns.boxplot( y="sepal_width",data=df,)
```

```python
print("Distribution of Petal Width")
sns.boxplot( y="petal_width",data=df,)
```
